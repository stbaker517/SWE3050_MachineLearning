```python
import numpy as np

def getLNorm(w,L):
    w=np.array(w)
    L1=[]
    L2=[]
    for _w in w:
        #print(_w)
        L1.append(np.sum(_w,axis=0))
        L2.append(np.linalg.norm(_w,2)**2)
    L1=np.array(L1)*L/2
    L2=np.array(L2)*L/2
    print("L1:",L1)
    print("L2:",L2)
    return L1,L2
    
```

```python
def sigmoid(x):
    x=np.array(x)
    return 1 / (1 +np.exp(-x))

def softmax(a): 
    a=np.array(a)
    c = np.max(a) 
    exp_a = np.exp(a) 
    sum_exp_a = np.sum(exp_a) 
    y = exp_a / sum_exp_a 
    return y
def SMratio(output):
    
    output=np.array(output)
    ret= softmax(output)
    s=np.sum(ret,axis=0)
    ret/=s
    return ret
```

```python
def batchNorm(net, ep, gamma, beta):
    net=np.array(net)
    m = np.mean(net) # mean
    v = np.var(net) # variance 
    d = np.sqrt(v) # standard deviation

    print(f"mean: {m}, variance: {v}, std: {d}")

    ### normalization
    
    net1 = []
    for idx in net:
        tmp = (idx-m)/np.sqrt(v + ep)
        net1.append(tmp)
    print("^net:",net1)

    ### gamma beta
    net2 = []
    for net in net1:
        mp = gamma * net + beta
        net2.append(tmp)
    print("~net:",net2)
    
    net1=np.array(net1)
    net2=np.array(net2)
    return net1,net2
```

```python
def leastSquare(x,y,poly):
    x=np.array(x)
    y=np.array(y).reshape(len(y),1)
    A=np.array([])
    A=[]
    for _x in x:
        #print(x)
        tmp=[1]
        for i in range(1,poly+1):
            tmp.append(_x**i)
        A.append(tmp)
    #print(tmp)
    A=np.array(A)

    At=A.transpose()

    AtA=np.dot(At,A)
    AtA_=np.linalg.inv(AtA)
   
    Aty=np.dot(At,y)
    W=np.dot(AtA_,Aty)
    W=np.round(W,3)
    return W
```

# API description
### 주의점: 모든 벡터형태는 np.array가 아닌 단순한 리스트로 집어넣을 것. (알아서 함수 내부에서 np.array로 바꿈)
#### 1. getLNorm(w,lambda) 
- return: (np.array) L1, L2
- w: np.array의 배열. [ w1 w2 w3 ...]
- lamba: lamda값.   
- example: getLNorm(np.array[ [1, 2, 3 ], [ 2, 3, 4 ]))

#### 2. sigmoid(x)
- return: (float) sigmoid(x)
- x: float

#### 3. softmax(x)
- return (float) softmex(x)
- x:float

#### 3-1: SMratio(outputs):  아웃풋들의 소프트맥스 비율을 알려주는 함수.
- return: exp(o_i) / sum(exp(o_k) 소프트맥스 비율을 알려줌.
- outputs: 아웃풋 밸류의 백터. np.array

#### 4: batchNorm(net, epsilon, gamma, beta)
- return: (np.array) ^net, ~net
- epsilon: deviation에 더해주는 상수
- gamma, beta: ~net 계산용

#### 5: leastSquare(x,y,poly):
- return: givendaya{x,y}의 최소제곱해 리턴. {w0, w1, w2, ... , wn}
- x: [ x1, x2, x3, ... ,xn]
- y: [ y1, y2 ,y3, ... ,yn]
- poly: 다항식의 최고차항

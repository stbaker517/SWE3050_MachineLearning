```python
import sklearn as sk
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

# Reading .csv

```python
data=pd.read_csv('./train.csv')
#data
```

# string -> int encoding

문자열을 정수형태로 바꾸어주는 이유는 학습 시에 문자열 혹은 문자 데이터를 입력시 학습 모델의 수식 적용이 불가능하기 때문이다.

```python
data_dropped=data.drop(['PassengerId' ,'Name', 'Ticket',  'Cabin'],axis=1)
data_dropped.loc[data_dropped.Sex=='male','Sex']=1
data_dropped.loc[data_dropped.Sex=='female','Sex']=0
data_dropped.loc[data_dropped.Embarked=='S','Embarked']=0
data_dropped.loc[data_dropped.Embarked=='Q','Embarked']=1
data_dropped.loc[data_dropped.Embarked=='C','Embarked']=2




#data_dropped
```

# split data into train and test

```python
dt_drop=data_dropped
dt_train=dt_drop[:600]
dt_test=dt_drop[600:]
```

```python
#dt_train
```

# Remove nan row

```python
dt_train=dt_train.dropna(axis=0)
dt_test=dt_test.dropna(axis=0)
```

```python
#dt_train
```

# Split X,Y

X는 입력값 Y는 출력값이다.

```python
X_train=dt_train.loc[:,['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
Y_train=dt_train.loc[:,['Survived']]

X_test=dt_test.loc[:,['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
Y_test=dt_test.loc[:,['Survived']]
```

```python
#X_train
```

# Data Normalization Z-score normalization

처음 normalize를 하지 않고 학습을 시킬 경우 regression의 경우에는 적은 iter에서 학습에 실패하였다.

그리고 knn와 DT의 경우에 아주 낮은 performance를 보여줬다.

처음에는 minmax 정규화를 시도했으나 생각보다 낮은 performance 증가 폭을 보여 Z score nomalization을 시도하였고 상당히 유의미한 performace 향상을 얻었다.

##### normalization을 하면서 중요한 점은 train data를 normalization을 할 때 사용한 mean값과 std값을 이용해 test데이터도 정규화 해주어야한다는 것이다.


```python
_max=X_train.loc[:,['Fare']].max()

_min=X_train.loc[:,['Fare']].min()
m=_max-_min


#Z score normalization
for col in X_train.columns:
    _std=X_train.loc[:,[col]].std()
    _mean=X_train.loc[:,[col]].mean()
    
    X_train.loc[:,[col]]=(X_train.loc[:,[col]]-_mean)/_std
    #X_train.loc[:,['Age']]/=50
    X_test.loc[:,[col]]=(X_test.loc[:,[col]]-_mean)/_std
    #X_test.loc[:,['Age']]/=50
```

# RESULT

### Max_depth =1

```python
md=1
model=sk.tree.DecisionTreeClassifier(criterion='entropy',max_depth=md)
model.fit(X_train,Y_train.values.ravel())
Y_pred=model.predict(X_test)
#print(Y_pred)
#print(Y_test.values.ravel())
ret=sk.metrics.accuracy_score(Y_test.values.ravel(),Y_pred)
F1=sk.metrics.f1_score(Y_test.values.ravel(),Y_pred)
print("Max_depth:",md,"Accuracy:",ret,"F1-score:",F1)
#fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(feature_names=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked'],decision_tree=model)
```

### Max_depth =2

```python
md=2
model=sk.tree.DecisionTreeClassifier(criterion='entropy',max_depth=md)
model.fit(X_train,Y_train.values.ravel())
Y_pred=model.predict(X_test)
#print(Y_pred)
#print(Y_test.values.ravel())
ret=sk.metrics.accuracy_score(Y_test.values.ravel(),Y_pred)
F1=sk.metrics.f1_score(Y_test.values.ravel(),Y_pred)
print("Max_depth:",md,"Accuracy:",ret,"F1-score:",F1)
fig = plt.figure(figsize=(20,10))
_ = tree.plot_tree(feature_names=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked'],decision_tree=model)
```

```python
md=3
model=sk.tree.DecisionTreeClassifier(criterion='entropy',max_depth=md)
model.fit(X_train,Y_train.values.ravel())
Y_pred=model.predict(X_test)
#print(Y_pred)
#print(Y_test.values.ravel())
ret=sk.metrics.accuracy_score(Y_test.values.ravel(),Y_pred)
F1=sk.metrics.f1_score(Y_test.values.ravel(),Y_pred)
print("Max_depth:",md,"Accuracy:",ret,"F1-score:",F1)
fig = plt.figure(figsize=(20,10))
_ = tree.plot_tree(feature_names=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked'],decision_tree=model)
```

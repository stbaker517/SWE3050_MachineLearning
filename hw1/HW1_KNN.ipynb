```python
data_dropped=data.drop(['PassengerId' ,'Name', 'Ticket',  'Cabin'],axis=1)
data_dropped.loc[data_dropped.Sex=='male','Sex']=1
data_dropped.loc[data_dropped.Sex=='female','Sex']=0
data_dropped.loc[data_dropped.Embarked=='S','Embarked']=0
data_dropped.loc[data_dropped.Embarked=='Q','Embarked']=1
data_dropped.loc[data_dropped.Embarked=='C','Embarked']=2




#data_dropped
```

```python
dt_train=dt_train.dropna(axis=0)
dt_test=dt_test.dropna(axis=0)
```

```python

Y_train=dt_train.loc[:,['Survived']]
X_train=dt_train.loc[:,['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
X_test=dt_test.loc[:,['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']]
Y_test=dt_test.loc[:,['Survived']]
```

```python

df=X_train
x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(data=x_scaled,columns=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked'])
X_train=df

```

```python
df=X_test
x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(data=x_scaled,columns=['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked'])
X_test=df
```

# RESULT

```python
for numOfNeighbors in range(2,6):
    print('K: ',numOfNeighbors)
 
    knn=sk.neighbors.KNeighborsClassifier(n_neighbors = numOfNeighbors)
    knn.fit(X_train,Y_train.values.ravel())
    Y_pred=knn.predict(X_test)
    #print(Y_pred)
    #print(Y_test.values.ravel())
    ret=sk.metrics.accuracy_score(Y_test.values.ravel(),Y_pred)
    F1=sk.metrics.f1_score(Y_test.values.ravel(),Y_pred)
    print('Accuracy: ', ret,'\nF1score: ',F1)
```

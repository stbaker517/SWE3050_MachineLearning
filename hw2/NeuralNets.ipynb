```python
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(1,1)
    def forward(self, x):
        x = self.fc1(x) # y=Ax+b
        return x
```

```python
net = Net()
print(net)
```

```python
print(list(net.parameters()))
```

```python
input = Variable(torch.randn(1,1,1), requires_grad=True) #1x1x1 -> one value
print(input)
```

```python
out = net(input)
print(out)
```

```python
import torch.optim as optim
def criterion(out, label):
    return (label - out)**2
optimizer = optim.SGD(net.parameters(), lr=0.01, momentum=0.5)
```

```python
data = [(1,3), (2,6), (3,9), (4,12), (5,15), (6,18)] # Ax + b will be that A=3 and b=0
```

```python
for epoch in range(100):
    for i, data2 in enumerate(data):
        X, Y = iter(data2)
        X, Y = Variable(torch.FloatTensor([X]), requires_grad=True), Variable(torch.FloatTensor([Y]), requires_grad=False)
        optimizer.zero_grad()
        outputs = net(X)
        loss = criterion(outputs, Y)
        loss.backward()
        optimizer.step()
        if (i % 10 == 0):
            print("Epoch {} - loss: {}".format(epoch, loss.data[0]))
```

```python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(1,10)
        self.fc2 = nn.Linear(10,1)
    def forward(self, x):
        x = self.fc2(self.fc1(x))
        return x
```
